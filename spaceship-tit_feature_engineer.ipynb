```python
import pandas as pd
df = pd.read_csv('ST_train_impute_final.csv')
```

```python
df.head()
```

```python
df[['deck', 'num', 'side']] = df['Cabin'].str.split('/', expand=True)
df[['Group', 'Member']] = df['PassengerId'].str.split('_', expand=True)
group_counts = df['Group'].value_counts()
df['Group_size'] = df['Group'].map(group_counts)
df['All_Zero'] =df[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].eq(0).all(axis=1)
df['Total_Spend'] = df[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)
df.head()
```

```python
df = df.drop(['PassengerId', 'Cabin', 'side','Name', 'Member'], axis=1)
df.head(11)
```

```python
df.to_csv('ST_train_feature_final.csv', index=False)
#df.to_csv('ST_test_feature_final.csv', index=False)
```
