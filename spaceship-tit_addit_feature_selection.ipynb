```python
import pandas as pd
df = pd.read_csv('ST_train.csv')
```

**HomePlanet**

```python
HomePlanet_feat_df = df[['HomePlanet', 'Transported']].copy()
HomePlanet_feat_df = HomePlanet_feat_df.dropna()
HomePlanet_feat_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
plt.figure(figsize=(7, 5))
sns.countplot(x='HomePlanet', hue='Transported', data=HomePlanet_feat_df, palette='mako')
plt.title('Transported Counts by HomePlanet')
plt.xlabel('HomePlanet')
plt.ylabel('Count')
plt.savefig('Transported Counts by HomePlanet.svg')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert 'Transported' to numeric (True to 1 and False to 0)
HomePlanet_feat_df['Transported'] = HomePlanet_feat_df['Transported'].astype(int)

# Apply label encoding to 'HomePlanet'
label_encoder = LabelEncoder()
HomePlanet_feat_df['HomePlanet_encoded'] = label_encoder.fit_transform(HomePlanet_feat_df['HomePlanet'])

# Calculate correlations
correlation_matrix = HomePlanet_feat_df[['HomePlanet_encoded', 'Transported']].corr()

print(correlation_matrix)
```

```python
# Convert 'Transported' to numeric (True to 1 and False to 0)
#HomePlanet_feat_df['Transported'] = HomePlanet_feat_df['Transported'].astype(int)

# One-hot encode 'HomePlanet'
df_encoded = pd.get_dummies(HomePlanet_feat_df, columns=['HomePlanet'])

# Calculate correlations
correlation_matrix = df_encoded.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.savefig('Transported HomePlanet correlation.svg')
plt.show()
```

```python
from sklearn.metrics import mutual_info_score
mi = mutual_info_score(HomePlanet_feat_df['HomePlanet_encoded'], HomePlanet_feat_df['Transported'])

print(f"Mutual Information Score: {mi}")
```
